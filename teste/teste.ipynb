```python

import matplotlib.pyplot as plt
import pandas as pd

plt.style.use('ggplot')
estado = [
    "AC", "AL", "AP", "AM", "BA", "CE", "DF", "ES", "GO", "MA", "MS", "MT", "MG", "PA", "PB", "PR", "PE", "PI", "RJ", "RN", "RS", "RO", "RR", "SC", "SP", "SE", "TO"
]
casos_primeiro_trim = [
    19625, 59725, 100940, 36468, 166154, 173882,
    106292, 83292, 67941, 120661, 127106, 24936,
    51865, 154685, 82794, 95005, 51477, 76112,
    165495, 50416, 38992, 32016, 66692, 84073,
    58713, 542304, 24824
]

casos_segundo_trim = [
    41620, 104818, 201013, 68201, 493400, 335091,
    251701, 248232, 308868, 200938, 542909, 133761,
    179427, 293540, 166484, 222166, 142672, 416566,
    434648, 118083, 95729, 68689, 449674, 492583,
    112505, 1462297, 90358
]

tamanho = len(casos_primeiro_trim)

casos_totais = []

for i in range(tamanho):
    soma = casos_primeiro_trim[i] + casos_segundo_trim[i]
    casos_totais.append(soma)

fig_a = plt.figure(figsize=(12,2.8))
ax = fig_a.add_subplot(111)
plt.title('Número de Casos em 2020 por Milhão')

df=pd.DataFrame({"estado":estado,"casos_totais":casos_totais})

plt.bar(df['estado'],df['casos_totais'],)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#desempregados_totais=[8389000,9249000,9878000,9974000] 
desempregados_mes=[2371000,1537000,1437000,1533000] 
desempregados_ano=[6018000,7712000,8441000,7644000] 

fig_c = plt.figure(figsize=(12.2,2.5))
ax = fig_c.add_subplot(111)

ax.set_xlabel('Trimestre de 2020',fontsize = 12)

xaxis = np.array(['1º','2º','3º','4º'])

STDV_desemprego_mes=np.std(desempregados_mes)
STDV_desemprego_ano=np.std(desempregados_ano)

color_desempregados_mes = 'red'
color_desempregados_ano = 'blue'

lineStyle_desempregados_mes={"linestyle":"-", "linewidth":2, "markeredgewidth":2,"elinewidth":2,"capsize":3}
lineStyle_desempregados_ano={"linestyle":"-", "linewidth":2, "markeredgewidth":2,"elinewidth":2,"capsize":3}

line_desempregados_mes=ax.errorbar(xaxis, desempregados_mes, yerr=STDV_desemprego_mes, **lineStyle_desempregados_mes, color=color_desempregados_mes, label='Desempregados Menos de Um Mês')
line_desempregados_ano=ax.errorbar(xaxis, desempregados_ano, yerr=STDV_desemprego_ano, **lineStyle_desempregados_ano, color=color_desempregados_ano, label='Desempregados Menos de Um Ano')

ax.legend()
```

```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

pessoas_ocupadas=[93115,84051,83439,87225] 
pessoas_desocupadas=[13148,13228,14598,14412] 

fig_d = plt.figure(figsize=(4,4))
ax = fig_d.add_subplot(111)

ax.set_ylabel('Relação entre pessoas com Ocupação e Desocupação',fontsize = 6,labelpad=20)
ax.set_xlabel('Trimestre de 2020',fontsize = 6)

xaxis = np.array(['1º','2º','3º','4º'])

STDV_pessoas_ocupadas=np.std(pessoas_ocupadas)
STDV_pessoas_desocupadas=np.std(pessoas_desocupadas)

color_pessoas_ocupadas = 'red'
color_pessoas_desocupadas = 'blue'

lineStyle_pessoas_ocupadas={"linestyle":"-", "linewidth":2, "markeredgewidth":2,"elinewidth":2,"capsize":3}
lineStyle_pessoas_desocupadas={"linestyle":"-", "linewidth":2, "markeredgewidth":2,"elinewidth":2,"capsize":3}

line_pessoas_ocupadas=ax.errorbar(xaxis, pessoas_ocupadas, yerr=STDV_pessoas_ocupadas, **lineStyle_pessoas_ocupadas, color=color_pessoas_ocupadas, label='Pessoas >=14 anos desocupadas (milhares)')
line_pessoas_desocupadas=ax.errorbar(xaxis, pessoas_desocupadas, yerr=STDV_pessoas_desocupadas, **lineStyle_pessoas_desocupadas, color=color_pessoas_desocupadas, label='Pessoas >=14 anos ocupadas (milhares)')


ax.legend()
```
